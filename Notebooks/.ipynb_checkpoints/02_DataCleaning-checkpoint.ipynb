```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raws/MUP_PHY_R26_P05_V10_D24_Prov.csv")

print(df.shape)
```

```python
df_clean = df.copy()
```

```python
duplicate_rows = df_clean.duplicated().sum()

print("Duplicate Rows:", duplicate_rows)
df_clean = df_clean.drop_duplicates()
print(df_clean.shape)
```

```python
duplicate_npi = df_clean.duplicated(subset="Rndrng_NPI").sum()

print("Duplicate NPIs:", duplicate_npi)
```

```python
df_clean.columns = (
    df_clean.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
        .str.replace("/", "_")
)
df_clean.columns
```

```python
missing_report = (
    df_clean.isnull()
    .sum()
    .sort_values(ascending=False)
)

missing_report[missing_report > 0].head(30)
```

```python
missing_percent = (
    df_clean.isnull()
    .mean() * 100
).sort_values(ascending=False)

missing_percent[missing_percent > 0].head(30)
```

```python
object_cols = df_clean.select_dtypes(include="object").columns

df_clean[object_cols] = df_clean[object_cols].fillna("Unknown")
df_clean[object_cols].isnull().sum().sum()
```

```python
df_clean.dtypes
```

```python
for col in object_cols:
    df_clean[col] = df_clean[col].str.strip()
```

```python
df_clean.isnull().sum().sum()
```

```python
missing_report = df_clean.isnull().sum().sort_values(ascending=False)
missing_report[missing_report > 0].head(40)
```

```python
missing_percent = (df_clean.isnull().mean() * 100).sort_values(ascending=False)
missing_percent[missing_percent > 0].head(40)
```

```python
numeric_cols = df_clean.select_dtypes(include="number").columns

for col in numeric_cols:
    missing_pct = df_clean[col].isnull().mean() * 100

    if missing_pct < 20:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
object_cols = df_clean.select_dtypes(include="object").columns

df_clean[object_cols] = df_clean[object_cols].fillna("Unknown")
```

```python
remaining_missing = (
    df_clean.isnull()
    .sum()
    .sort_values(ascending=False)
)

remaining_missing[remaining_missing > 0].head(30)
```

```python
missing_df = pd.DataFrame({
    "Missing_Count": df_clean.isnull().sum(),
    "Missing_Percent": round(df_clean.isnull().mean() * 100, 2)
})

missing_df = missing_df[missing_df["Missing_Count"] > 0]
missing_df = missing_df.sort_values("Missing_Percent", ascending=False)

missing_df
```

```python
missing_df.to_csv("../Reports/missing_value_report.csv", index=True)
```

```python
high_missing = missing_df[missing_df["Missing_Percent"] > 50]
medium_missing = missing_df[
    (missing_df["Missing_Percent"] >= 20) &
    (missing_df["Missing_Percent"] <= 50)
]
low_missing = missing_df[missing_df["Missing_Percent"] < 20]

print("High Missing:", len(high_missing))
print("Medium Missing:", len(medium_missing))
print("Low Missing:", len(low_missing))
```

```python
numeric_cols = df_clean.select_dtypes(include="number").columns

for col in numeric_cols:
    if df_clean[col].isnull().mean() < 0.20:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
object_cols = df_clean.select_dtypes(include="object").columns

for col in object_cols:
    df_clean[col] = df_clean[col].fillna("Unknown")
```

```python
remaining = (
    df_clean.isnull()
            .sum()
            .sort_values(ascending=False)
)

remaining[remaining > 0]
```

```python
print("=" * 60)
print("DATA QUALITY REPORT")
print("=" * 60)

print(f"Original Rows: {df.shape[0]:,}")
print(f"Original Columns: {df.shape[1]}")

print()

print(f"Duplicate Rows: {df_clean.duplicated().sum():,}")
print(f"Duplicate NPIs: {df_clean.duplicated(subset='rndrng_npi').sum():,}")

print()

print(f"Columns with Missing Values: {missing_df.shape[0]}")
print(f"Total Missing Values: {df_clean.isnull().sum().sum():,}")
```

```python
import os

# Create Reports folder if it doesn't exist
os.makedirs("../Reports", exist_ok=True)

# Save the report
missing_df.to_csv("../Reports/missing_value_report.csv", index=True)

print("Missing Value Report saved successfully!")
```
