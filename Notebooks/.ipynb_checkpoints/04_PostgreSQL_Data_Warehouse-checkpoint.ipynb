```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/processed/clean_provider_data.csv")

print(df.shape)
df.head()
```

```python
engine = create_engine(
    "postgresql://postgres:coolcoconut2210@localhost:5432/healthcare_fraud_db"
)

print("Connected Successfully!")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("SELECT version();"))
    print(result.fetchone())
```

```python
df.to_sql(
    "providers",
    engine,
    if_exists="replace",
    index=False,
    chunksize=50000
)

print("Providers table loaded successfully!")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("SELECT COUNT(*) FROM providers;"))
    print(result.fetchone())
```

```python
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM providers LIMIT 5;"))
    for row in result:
        print(row)
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT table_name
        FROM information_schema.tables
        WHERE table_schema='public';
    """))

    for table in result:
        print(table[0])
```

```python
with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT column_name, data_type
        FROM information_schema.columns
        WHERE table_name='providers';
    """))

    for row in result:
        print(row)
```
