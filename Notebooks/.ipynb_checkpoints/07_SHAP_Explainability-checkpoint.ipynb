```python
import pandas as pd
import numpy as np
import shap
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)
```

```python
features = pd.read_csv("../data/processed/provider_risk_scores.csv")

print(features.shape)
features.head()
```

```python
model_features = [
    "payment_per_beneficiary",
    "services_per_beneficiary",
    "charge_to_payment_ratio",
    "allowed_to_charge_ratio",
    "standardized_payment_ratio",
    "payment_vs_state_avg",
    "payment_vs_specialty_avg",
    "tot_mdcr_pymt_amt",
    "tot_srvcs",
    "tot_benes"
]

X = features[model_features]
X = X.fillna(X.median())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = IsolationForest(
    n_estimators=200,
    contamination=0.02,
    random_state=42
)

model.fit(X_scaled)

print("Model trained for SHAP explanation.")
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=model_features)

sample = X_scaled_df.sample(1000, random_state=42)

print(sample.shape)
```

```python
explainer = shap.TreeExplainer(model)

shap_values = explainer.shap_values(sample)

print("SHAP values calculated.")
```

```python
shap.summary_plot(
    shap_values,
    sample,
    feature_names=model_features,
    show=True
)
```

```python
plt.figure()

shap.summary_plot(
    shap_values,
    sample,
    feature_names=model_features,
    show=False
)

plt.savefig(
    "../Images/shap_summary.png",
    dpi=300,
    bbox_inches="tight"
)

print("SHAP summary image saved.")
```

```python
top_provider = (
    features.sort_values("fraud_risk_score", ascending=False)
    .head(1)
)

top_provider[
    [
        "rndrng_npi",
        "rndrng_prvdr_last_org_name",
        "rndrng_prvdr_first_name",
        "rndrng_prvdr_type",
        "rndrng_prvdr_state_abrvtn",
        "fraud_risk_score"
    ]
]
```

```python
shap_importance = pd.DataFrame({
    "feature": model_features,
    "mean_abs_shap": np.abs(shap_values).mean(axis=0)
}).sort_values("mean_abs_shap", ascending=False)

shap_importance
```

```python
shap_importance.to_csv(
    "../data/processed/shap_feature_importance.csv",
    index=False
)

print("SHAP feature importance saved.")
```
