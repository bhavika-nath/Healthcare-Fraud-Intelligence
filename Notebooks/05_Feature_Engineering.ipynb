```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
features = pd.read_csv("../data/processed/clean_provider_data.csv")

print(features.shape)
features.head()
```

```python
features["payment_per_beneficiary"] = (
    features["tot_mdcr_pymt_amt"] /
    features["tot_benes"]
)

# Services per beneficiary
features["services_per_beneficiary"] = (
    features["tot_srvcs"] /
    features["tot_benes"]
)

# Submitted charge to Medicare payment ratio
features["charge_to_payment_ratio"] = (
    features["tot_sbmtd_chrg"] /
    features["tot_mdcr_pymt_amt"]
)

# Medicare allowed amount to submitted charge ratio
features["allowed_to_charge_ratio"] = (
    features["tot_mdcr_alowd_amt"] /
    features["tot_sbmtd_chrg"]
)

# Standardized payment ratio
features["standardized_payment_ratio"] = (
    features["tot_mdcr_stdzd_amt"] /
    features["tot_mdcr_pymt_amt"]
)

# Replace infinite values
features.replace([np.inf, -np.inf], np.nan, inplace=True)

print("Core fraud features created.")
```

```python
state_avg_payment = (
    features.groupby("rndrng_prvdr_state_abrvtn")["tot_mdcr_pymt_amt"]
            .mean()
            .reset_index()
)

state_avg_payment.columns = [
    "rndrng_prvdr_state_abrvtn",
    "state_avg_payment"
]

features = features.merge(
    state_avg_payment,
    on="rndrng_prvdr_state_abrvtn",
    how="left"
)

print("State averages added.")
```

```python
specialty_avg_payment = (
    features.groupby("rndrng_prvdr_type")["tot_mdcr_pymt_amt"]
            .mean()
            .reset_index()
)

specialty_avg_payment.columns = [
    "rndrng_prvdr_type",
    "specialty_avg_payment"
]

features = features.merge(
    specialty_avg_payment,
    on="rndrng_prvdr_type",
    how="left"
)

print("Specialty averages added.")
```

```python
features["payment_vs_state_avg"] = (
    features["tot_mdcr_pymt_amt"] /
    features["state_avg_payment"]
)

features["payment_vs_specialty_avg"] = (
    features["tot_mdcr_pymt_amt"] /
    features["specialty_avg_payment"]
)

print("Peer comparison features created.")
```

```python
payment_threshold = features["tot_mdcr_pymt_amt"].quantile(0.95)
service_threshold = features["tot_srvcs"].quantile(0.95)

features["high_payment_flag"] = (
    features["tot_mdcr_pymt_amt"] >= payment_threshold
).astype(int)

features["high_service_flag"] = (
    features["tot_srvcs"] >= service_threshold
).astype(int)

print("Risk flags created.")
```

```python
new_features = [
    "payment_per_beneficiary",
    "services_per_beneficiary",
    "charge_to_payment_ratio",
    "allowed_to_charge_ratio",
    "standardized_payment_ratio",
    "payment_vs_state_avg",
    "payment_vs_specialty_avg",
    "high_payment_flag",
    "high_service_flag"
]

features[new_features].describe()
```

```python
features.to_csv(
    "../data/processed/provider_features.csv",
    index=False
)

print("Feature engineering completed successfully!")
```
