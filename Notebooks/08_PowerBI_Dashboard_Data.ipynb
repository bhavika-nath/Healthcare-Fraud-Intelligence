```python
import pandas as pd
import numpy as np
import os

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/processed/provider_risk_scores.csv")

print(df.shape)
df.head()
```

```python
def risk_category(score):
    if score >= 80:
        return "High Risk"
    elif score >= 50:
        return "Medium Risk"
    else:
        return "Low Risk"

df["risk_category"] = df["fraud_risk_score"].apply(risk_category)
```

```python
dashboard_df = df[
    [
        "rndrng_npi",
        "rndrng_prvdr_last_org_name",
        "rndrng_prvdr_first_name",
        "rndrng_prvdr_type",
        "rndrng_prvdr_state_abrvtn",
        "tot_benes",
        "tot_srvcs",
        "tot_sbmtd_chrg",
        "tot_mdcr_pymt_amt",
        "payment_per_beneficiary",
        "services_per_beneficiary",
        "charge_to_payment_ratio",
        "payment_vs_state_avg",
        "payment_vs_specialty_avg",
        "fraud_risk_score",
        "risk_category"
    ]
]
```

```python
state_risk_summary = (
    dashboard_df.groupby("rndrng_prvdr_state_abrvtn")
    .agg(
        total_providers=("rndrng_npi", "nunique"),
        avg_risk_score=("fraud_risk_score", "mean"),
        high_risk_providers=("risk_category", lambda x: (x == "High Risk").sum()),
        total_medicare_payment=("tot_mdcr_pymt_amt", "sum"),
        total_services=("tot_srvcs", "sum")
    )
    .reset_index()
)
```

```python
specialty_risk_summary = (
    dashboard_df.groupby("rndrng_prvdr_type")
    .agg(
        total_providers=("rndrng_npi", "nunique"),
        avg_risk_score=("fraud_risk_score", "mean"),
        high_risk_providers=("risk_category", lambda x: (x == "High Risk").sum()),
        total_medicare_payment=("tot_mdcr_pymt_amt", "sum"),
        total_services=("tot_srvcs", "sum")
    )
    .reset_index()
)
```

```python
os.makedirs("../Dashboards/powerbi_data", exist_ok=True)

dashboard_df.to_csv("../Dashboards/powerbi_data/provider_dashboard_data.csv", index=False)
state_risk_summary.to_csv("../Dashboards/powerbi_data/state_risk_summary.csv", index=False)
specialty_risk_summary.to_csv("../Dashboards/powerbi_data/specialty_risk_summary.csv", index=False)

print("Power BI dashboard datasets saved successfully.")
```
